```python
import os
from openai import OpenAI
from ollama import Client as OllamaClient
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY")) 
ollama_client = OllamaClient(host="http://localhost:11434")
```

Initialize conversation state

```python
conversation = [
    {
        "role": "system",
        "content": "You are a helpful customer service assistant for a grocery delivery service called ZimGroceries. You only answer questions about groceries and deliveries."
    }
]
```

Send openai client user message

```python
def ask_openai(message,conversation):
    conversation.append({"role": "user", "content": message})
    response = client.responses.create(
        model = "gpt-4o-mini",
        input = conversation
    )
    conversation.append({"role": "assistant", "content": response.output_text})
    return response.output_text
```

send ollama message

```python
def ask_ollama(message, conversation):
    conversation.append({"role": "user", "content": message})

    response = ollama_client.chat(
        model="llama3.2:3b",
        messages=conversation
    )

    reply = response["message"]["content"]

    conversation.append({"role": "assistant", "content": reply})
    return reply
```

```python
# Example usage:
user_message = "What are your delivery hours?"
print("OpenAI Response:")
print(ask_openai(user_message, conversation))
print("\nOllama Response:")
print(ask_ollama(user_message, conversation))

user_message = "How to make money from home with high speed internet?"
print("\nOpenAI Response:")
print(ask_openai(user_message, conversation))
print("\nOllama Response:")
print(ask_ollama(user_message, conversation))
```
